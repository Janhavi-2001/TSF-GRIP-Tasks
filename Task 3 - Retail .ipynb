The value count and 'Quantity' are two DIFFERENT parameters analysed in this EDA. The value count represents the transactions made and the 'Quantity' represents the amount of a particular commodity bought for that particular transaction!

### Step 3: Exploratory Data Analysis

Since the dataset contains information about the superstore branches only in the United States, the 'Country' column can be dropped as a result of redundancy. Morover, we won't be requiring the postal code, since the analysis is to be done at a city level.

```python
data = data.drop(columns = ['Country', 'Postal Code'])
```

In order to get a clear idea of the overall profits made, sales prices and discounts offered for each section, we will first need to understand their trends as a whole. Accordingly, we would be able to derive any conclusions for the following sections respectively. We can check out how the data is correlated using .corr().

```python
stats = data.corr()
```

```python
stats
```

```python
sns.lineplot(x = 'Sales', y = 'Profit', data = stats)
plt.xticks(size = 12)
plt.yticks(size = 12)
plt.title('Correlation between Sales and Profits', size = 15)
plt.xlabel('Sales', size = 13)
plt.ylabel('Profit', size = 13)
```

From the plot above, it is implicable that there was an initial rise in profits as the sales grew, only to drop sharply after the sales increased (at 0.5).

### I)  'Ship Mode' section

```python
shipping_mode = data.groupby('Ship Mode').size()
```

```python
shipping_mode
```

```python
fig, ax = plt.subplots(figsize = (8,8))
explode = (0.0, 0.0, 0.0, 0.1)
shipping_mode.plot.pie(shadow = True, explode = explode, fontsize = 15, autopct = "%1.2f%%", startangle = 45)
plt.title('Percentage Shares of Shipping Mode', fontsize = 15)
```

We can conclude from the above pie graph, that a large part of the shipping took place under the **'Standard Class'** section, followed by the **'Second Class'**, **'First Class'** and **'Same day'** sections respectively.

### Ship Mode Quantity of Purchase

```python
quant_ship = data.groupby('Ship Mode')['Quantity'].sum().sort_values(ascending = False)
```

```python
quant_ship
```

```python
quant_ship.plot(kind = 'bar', color = 'pink', figsize = (10,7))
plt.xticks(size = 14)
plt.yticks(size = 14)
sns.set_style('whitegrid')
plt.xlabel('Ship Mode', size = 15)
plt.ylabel('Revenue', size = 15)
plt.title('Ship-Mode Wise Quantity of Purchase', size = 18)
```

The maximum quantity of purchase was done under the **'Standard Class'**, followed by the **'Second Class'**, the **'Third Class'**, and the **'Same Day Class'** respectively. This aligns with the number of transactions done within each 'Shipping Mode'.

### Ship Mode Sales and Profit Analysis

```python
ship_analysis = data.groupby('Ship Mode')[['Sales', 'Profit']].sum().sort_values(by = 'Sales', ascending = False)
```

```python
ship_analysis
```

```python
ship_analysis.plot(kind = "bar", color = ['Blue', 'Orange'], figsize = (10,7))
plt.xticks(size = 14)
plt.yticks(size = 14)
sns.set_style('whitegrid')
plt.xlabel('Ship Mode', size = 15)
plt.ylabel('Revenue', size = 15)
plt.title('Ship-Mode Wise Revenue Generation Analysis', size = 18)
```

It is observed that **as the sales revenue grew for each Shipping Mode, the profits increased as well**. The profits generated are **directly proportional** to the sales revenue generated. Moreover, the maximum sales revenue and profit generated is **at par** with the number of transactions for each 'Shipping Mode'.

## What can be concluded from the above graphs?

## From the above analysis, we can conclude that:

## The relation between the 'Shipping Mode' and the 'Segment' with respect to the 'Sales' revenue and 'Profits' generated, shows a positive correlation. The same applies with the 4 regions as well.

## The 'Consumer' class was responsible for the most number of purchases made. Additionally, the greatest source of sales revenue and profits, were from this segment of buyers.

## In contrast to the quantity of purchase, wherein **'Office Supplies'** took the lead; the **'Technology'** and **'Furniture'** categories generated more sales revenue due to greater costs of the goods under these categories. Speaking of profits, the 'Furniture' category made the least profits, inspite of having generated a good sales revenue. This could be due to a lesser demand for furniture items, from the customers.

## The 'Technology' category made considerable profits in all regions in the US, and this could be attributed to the large demand for electronic gadgets. The 'Office Supplies' has shown a steady sales revenue amount altogether.

## The 'Southern Region' and the 'Central Region' lag behind in sales revenue and profits generated as compared to the 'Western Region' and the 'Eastern Region'. This could be due to a lesser number of outlets in these regions, or maybe even inadequate marketing in the cities within these areas.
